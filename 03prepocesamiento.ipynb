2. Normalización

```python
df['precleaned'] = df['review'].apply(lambda x: clear_text(x).lower())

print(df['precleaned'].iloc[0])
```

3. Eliminación de Stop Words

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words_en = set(stopwords.words('english'))

def tokenize_and_remove_stopwords(doc):
   
    tokens = word_tokenize(doc)
  
    filtered_tokens = [word for word in tokens if word not in stop_words_en and word.isalpha()]
    return filtered_tokens

#preprocesamiento
df['pretokens'] = df['precleaned'].apply(tokenize_and_remove_stopwords)

print("Ejemplo:")
print(df['pretokens'].iloc[0][:20])
```

4. Stemming

```python
from nltk.stem import porter
stemmer = porter.PorterStemmer()
```

```python
def apply_stemming(tokens):
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    return stemmed_tokens


df['stemmed'] = df['pretokens'].apply(apply_stemming)


print(df['stemmed'].iloc[0][:20])
```

5. Comprobación


```python
def get_all_tokens(series):
    all_tokens = series.sum()
    return all_tokens
```

```python
def get_all_tokens(series):
    all_tokens = series.sum()
    return all_tokens

sin_stemming = get_all_tokens(df['pretokens'])
vocab_sin = len(set(sin_stemming))
con_stemming = get_all_tokens(df['stemmed'])
vocab_con = len(set(con_stemming))

print(f"**Vocabulario si stemming:** {vocab_sin:,}")
print(f"**Vocabulario con stemming:** {vocab_con:,}")
```
